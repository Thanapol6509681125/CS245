# Calculate statistical values for all variables

```python
import statistics
```

```python
print("Mean of sepal length = ", statistics.mean(iris_dataset["sepal length"]))
print("Mean of sepal width = ", statistics.mean(iris_dataset["sepal width"]))
print("Mean of petal length = ", statistics.mean(iris_dataset["petal length"]))
print("Mean of petal width = ", statistics.mean(iris_dataset["petal width"]))
#Mean
```

```python
print("Median of sepal length = ", statistics.median(iris_dataset["sepal length"]))
print("Median of sepal width = ", statistics.median(iris_dataset["sepal width"]))
print("Median of petal length = ", statistics.median(iris_dataset["petal length"]))
print("Median of petal width = ", statistics.median(iris_dataset["petal width"]))
#Median
```

```python
print("Mode of sepal length = ", statistics.mode(iris_dataset["sepal length"]))
print("Mode of sepal width = ", statistics.mode(iris_dataset["sepal width"]))
print("Mode of petal length = ", statistics.mode(iris_dataset["petal length"]))
print("Mode of petal width = ", statistics.mode(iris_dataset["petal width"]))
# Mode
```

```python
print("Standard derivation of sepal length = ", statistics.stdev(iris_dataset["sepal length"]))
print("Standard derivation of sepal width = ", statistics.stdev(iris_dataset["sepal width"]))
print("Standard derivation of petal length = ", statistics.stdev(iris_dataset["petal length"]))
print("Standard derivation of petal width = ", statistics.stdev(iris_dataset["petal width"]))
# Standard derivation
```

```python
print("Skewness of sepal length = ", iris_dataset["sepal length"].skew())
print("Skewness of sepal width = ", iris_dataset["sepal width"].skew())
print("Skewness of petal length = ", iris_dataset["petal length"].skew())
print("Skewness of petal width = ", iris_dataset["petal width"].skew())
# Skewness
```

```python
print("Kurtosis of sepal length = ", iris_dataset["sepal length"].kurt())
print("Kurtosis of sepal width = ", iris_dataset["sepal width"].kurt())
print("Kurtosis of petal length = ", iris_dataset["petal length"].kurt())
print("Kurtosis of petal width = ", iris_dataset["petal width"].kurt())
# Kurtosis
```

# Calculate covariance and correlation values between petal length and petal width

```python
df = iris_dataset[["petal length", "petal width"]]
df.cov()
# Covariance
```

```python
df2 = iris_dataset[["petal length", "petal width"]]
df2.corr()
# Correlation
```

# What do the covariance and correlation tell about the relation between petal length and petal width

Covariance shows if petal length and width change together (positive or negative), while correlation measures how strongly they do so (closer to -1 or +1 means stronger relationship).
